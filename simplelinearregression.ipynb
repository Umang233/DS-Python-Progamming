```python
#import  libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"]=(10,5)
plt.rcParams["figure.dpi"]=250
sns.set_style("darkgrid")
```

```python
#read data
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/NewspaperData.csv')
df
```

```python
df.describe()
```

```python
#related columns
data=df[["daily","sunday"]]
```

```python
#correlation analyasis
data.corr()
```

```python
sns.heatmap(data.corr(),annot=True)# annot is use to mark on the columns
```

```python
#import statmodeel
import statsmodels.formula.api as smf
```

```python
model=smf.ols("sunday~daily",data=data).fit()
#create model
```

```python
#model parameter
model.params
```

```python
#line equation 
daily=650
y=1.339715*daily+13.835630
y
```

```python
model.pvalues
```

```python
model.rsquared
```

### Prediction for new data

```python
newdf=pd.DataFrame({"daily":[120,520,850,350,452]})
newdf
```

```python
model.predict(newdf)
```
